```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
```

```python
# Upload image file for transformation
image_1 = files.upload()
```

```python
image_path = list(image_1.keys())[0]

# Read image (BGR → RGB)
img = cv.imread(image_path)
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

# Normalize image to [0,1]
img_norm = img / 255.0
```

```python
#0.5 gamma correction
img_gamma_half = np.power(img_norm, 0.5)

# 2 gamma correction
img_gamma_two = np.power(img_norm, 2)
```

```python
# show image results
plt.figure(figsize=(10, 10))
plt.subplot(1, 3, 1)
plt.imshow(img)
plt.title('Original image')
plt.axis('off')
plt.subplot(1, 3, 2)
plt.imshow(img_gamma_half)
plt.title('0.5 gamma correction')
plt.axis('off')
plt.subplot(1, 3, 3)
plt.imshow(img_gamma_two)
plt.title('2 gamma correction')
plt.axis('off')
plt.show()
```
