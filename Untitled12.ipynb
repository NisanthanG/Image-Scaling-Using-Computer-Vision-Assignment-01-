```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
```

```python
# Upload image file for transformation
image_1 = files.upload()
```

```python
image_path = list(image_1.keys())[0]

# Read image (BGR → RGB)
img = cv.imread(image_path)
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

# Normalize image to [0,1]
img_norm = img / 255.0
```

```python
#0.5 gamma correction
img_gamma_half = np.power(img_norm, 0.5)

# 2 gamma correction
img_gamma_two = np.power(img_norm, 2)
```

```python
# show image results
plt.figure(figsize=(10, 10))
plt.subplot(1, 3, 1)
plt.imshow(img)
plt.title('Original image')
plt.axis('off')
plt.subplot(1, 3, 2)
plt.imshow(img_gamma_half)
plt.title('0.5 gamma correction')
plt.axis('off')
plt.subplot(1, 3, 3)
plt.imshow(img_gamma_two)
plt.title('2 gamma correction')
plt.axis('off')
plt.show()
```

```python
#  img_norm has the normalized image to [0,1]
#  Contrast Stretching (linear piecewise transformation)
stretched_img = np.zeros_like(img_norm)
r1 = 0.2
r2 = 0.8
# apply r1 = 0.2
stretched_img[img_norm < r1] = 0
mask = (img_norm >= r1) & (img_norm < r2)
stretched_img[mask] = (img_norm[mask] - r1) / (r2 - r1)
stretched_img[img_norm >= r2] = 1

#show results
plt.figure(figsize=(10, 10))
plt.subplot(1,2,1)
plt.imshow(img)
plt.title('Original image')
plt.axis('off')
plt.subplot(1,2,2)
plt.imshow(stretched_img)
plt.title('Contrast Stretching')
plt.axis('off')
plt.show()
```

```python
# Load image
img = cv.imread(list(image_2.keys())[0])
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

# Convert to LAB
lab = cv.cvtColor(img, cv.COLOR_RGB2LAB)
L, A, B = cv.split(lab)

# Normalize L channel
L_norm = L / 255.0

# Gamma correction
gamma = 0.5   # change to 2.0 if needed
L_gamma = np.power(L_norm, gamma)

# Convert back
L_new = np.uint8(L_gamma * 255)
lab_new = cv.merge((L_new, A, B))
img_corrected = cv.cvtColor(lab_new, cv.COLOR_LAB2RGB)

# Plot histograms
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.hist(L.ravel(), bins=256)
plt.title("Original L* Histogram")

plt.subplot(1,2,2)
plt.hist(L_new.ravel(), bins=256)
plt.title("Gamma Corrected L* Histogram")

plt.show()
```

```python
# Function to to equalize the histogram of an image

def my_hist_equalization(image):
    # Flatten image
    flat = image.flatten()

    # Step 1: Histogram
    hist = np.zeros(256)
    for pixel in flat:
        hist[pixel] += 1

    # Step 2: Normalize histogram (PDF)
    pdf = hist / np.sum(hist)

    # Step 3: CDF
    cdf = np.zeros(256)
    cdf[0] = pdf[0]
    for i in range(1, 256):
        cdf[i] = cdf[i-1] + pdf[i]

    # Step 4: Mapping
    cdf_normalized = np.round(cdf * 255).astype(np.uint8)

    # Step 5: Apply mapping
    equalized = cdf_normalized[image]

    return equalized
```

```python
# Apply the program to the image
equalized = my_hist_equalization(image_1)
```

```python
foreground = cv.bitwise_and(gray, gray, mask=mask)
```
